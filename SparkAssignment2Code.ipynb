```
nonEmptyRDD = myRDD.filter(lambda line:len(line.strip())>0)
```

```
splitRDD = nonEmptyRDD.map(lambda line:line.split(' ',1))
```

```
testRDD = splitRDD.map(lambda pair:(pair[0],pair[1].lower()))
```

```
testRDD.first()
```

```
def removePunc(verse):
  punc = [',','.',':',';','!','(',')','?']
  for i in punc:
    verse = verse.replace(i,' ')
  return set(verse.split(' '))
def f(x):
  return x
def events(e):
  returnlist = []
  for ee in e:
    returnlist.append(ee)
  return returnlist
```

```
testRDD2 = testRDD.mapValues(removePunc)
```

```
testRDD2.first()
```

```
testRDD2 = testRDD2.flatMapValues(f)
testRDD2 = testRDD2.filter(lambda (x,y):y!='')
```

```
testRDD2 = testRDD2.map(lambda (x,y):(y,x))
```

```
gbk = testRDD2.groupByKey()
testRDD3 = gbk.mapValues(events)
```

```
testRDD3= testRDD3.sortByKey()
```

```
testRDD3.take(2617)
```

```
testRDD4 = testRDD3.map(lambda (x,y):(x,(len(y),y)))
testRDD4 = testRDD4.sortBy(lambda (x,y):y[0],False)
```

```
testRDD4 = testRDD4.map(lambda (x,y):(x,y[1]))
```
